```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import style
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('Placement_Data_Full_Class.csv')
```

```python
df.head()
```

```python
df.isnull().sum().sort_values(ascending = False)
```

```python
df['salary'] = df['salary'].fillna(df['salary'].mean())
```

```python
df.head()
```

```python
df.isnull().sum().sort_values(ascending = False)
```

```python
fig = plt.figure(figsize=(18,18))
gs = fig.add_gridspec(3,4)
gs.update(wspace=0.3, hspace=0.15)
ax0 = fig.add_subplot(gs[0,1])
background_color = "#FFF8DC"
sns.countplot(ax=ax0, data=df, x='gender',palette = 'Oranges', hue='status',edgecolor='black',**{'hatch':'/','linewidth':2})
```

```python
fig = plt.figure(figsize=(18,18))
gs = fig.add_gridspec(3,4)
gs.update(wspace=0.3, hspace=0.15)
ax1 = fig.add_subplot(gs[0,1])
sns.countplot(ax=ax1, data=df, x='hsc_b',palette = 'Oranges',hue='status',edgecolor='black',**{'hatch':'/','linewidth':2})
```

```python
fig = plt.figure(figsize=(18,18))
gs = fig.add_gridspec(3,4)
gs.update(wspace=0.3, hspace=0.15)
ax2 = fig.add_subplot(gs[0,1])
sns.countplot(ax=ax2, data=df, x='hsc_s',palette = 'Oranges',hue='status',edgecolor='black',**{'hatch':'/','linewidth':2})
```

```python
fig = plt.figure(figsize=(18,18))
gs = fig.add_gridspec(3,4)
gs.update(wspace=0.3, hspace=0.15)
ax3 = fig.add_subplot(gs[0,1])
sns.countplot(ax=ax3, data=df, x='degree_t',palette = 'Oranges',hue='status',edgecolor='black',**{'hatch':'/','linewidth':2})
```

```python
#show the relation between salary and students as the count of students decreases when salary increases

fig = plt.figure(figsize=(22,15))
gs = fig.add_gridspec(3,3)
gs.update(wspace=0.3, hspace=0.15)
ax1 = fig.add_subplot(gs[0,1])
sns.histplot(ax=ax1, data=df[df['salary']>0], x='salary',palette = 'RdPu', kde=True, edgecolor='black')
```

```python
#salary analysis of placed students via specialisation

fig = plt.figure(figsize=(25,25))
gs = fig.add_gridspec(3,3)
gs.update(wspace=0.3, hspace=0.15)
ax1 = fig.add_subplot(gs[0,1])
sns.swarmplot(ax=ax1,y = df['salary'], x = df['specialisation'])
```

```python
#salary analysis for placed students male or female

fig = plt.figure(figsize=(25,25))
gs = fig.add_gridspec(3,3)
gs.update(wspace=0.3, hspace=0.15)
ax2 = fig.add_subplot(gs[0,1])
sns.swarmplot(y = df['salary'], x = df['gender'],palette='CMRmap_r')
```

```python
fig, axs = plt.subplots(ncols=3,figsize=(20,5))
sns.scatterplot(x = 'degree_p',y='hsc_p',hue='status',data = df, ax= axs[0])
sns.scatterplot(x = 'degree_p',y='hsc_p',hue='gender',data = df, ax= axs[1], palette="muted")
sns.scatterplot(x = 'degree_p',y='hsc_p',hue='degree_t',data = df, palette="dark", ax= axs[2])
```
